## Drawing geometrical shapes on an image
Let's take an image of the "wall of love" in Paris. It’s a wall which is filled with the words “I love you” in all kinds of international languages.

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

#read the image
img=mpimg.imread('walloflove.jpg')
imgplot = plt.imshow(img)
plt.show()
```

# Drawing lines
Next, we’ll draw a green line.
If you look carefully at the method parameters and compare them to that of the rectangle, you’ll notice that they are identical:

```
# Copy the image
img_copy = img.copy()

# Draw a line
cv2.line(img_copy, (60, 20), (500, 300), (255, 255, 0), 5)
#last argument is thickness expressed in dots; 1 is the smallest value

# Draw a arrowed line
cv2.arrowedLine(img_copy, (500, 300), (600, 200), (100, 200, 100), 20)
plt.imshow(img_copy)
plt.show()
```

# Drawing rectangles
What we’re going to do is finding the words in our language and marking them with a rectangle. We’ll look up for the Italian text "Ti amo". Before we move on with drawing on an image with OpenCV, take note that drawing operations on images are performed in-place. Therefore at the beginning of each code block, we make a copy of the original image storing the copy as 'img_copy'. We then proceed to draw on the image called 'img_copy' in-place so we do not destroy our original image. Let's try to draw a rectangle with `cv2.rectangle()`. We need to give the coordinates values for the upper left point and the lower right point.

```
# Copy the image
img_copy = img.copy()
# Draw an empty rectangle
#pt1 is the upper left vertex, pt2 is the lower right vertex
cv2.rectangle(img_copy, pt1 = (750, 450), pt2 = (850, 490), color = (255, 0, 0), thickness = 5)

# Draw a filled-in rectangle (thickness=-1)
cv2.rectangle(img_copy, (200, 200), (300, 400), (0, 255, 0), -1)
plt.imshow(img_copy)
plt.show()
```

# Drawing circles
Now let's draw a blue circle around the  heart on the left in the wall (can you see that?). With `cv2.circle()`, we need to specify the point of its center and the length of its radius.

```
# Copy the image
img_copy = img.copy()
# Draw an empty circle
cv2.circle(img_copy, center = (870, 300), radius = 40, color = (255, 255, 0), thickness = 5)

# Draw a filled-in circle
cv2.circle(img_copy, center = (500, 300), radius = 40, color = (0, 255, 255), thickness = -1)

plt.imshow(img_copy)
plt.show()
```

# Drawing ellipses
To draw an ellipse, we use the `cv2.ellipse()` that takes the following arguments:

Image object on which the ellipse needs to be drawn

Center pixel’s coordinates

Length of the minor and major axes

Rotation angle of the ellipse (calculated
counterclockwise)

Starting angle (calculated clockwise)

Final angle (calculated clockwise)

Color

Thickness

```
# Copy the image
img_copy = img.copy()
# Draw a yellow ellipse
center_coordinates = (120, 100)
axesLength = (100, 50)
angle = 20
startAngle = 0
endAngle = 360
color = (255, 255, 0)
# Line thickness of 5 px
thickness = 5
image = cv2.ellipse(img_copy, center_coordinates, axesLength,
           angle, startAngle, endAngle, color, thickness)

#add another ellipse (green, filled in)
# Line thickness of -1 px
thickness = -1
center_coordinates = (220, 300)
axesLength = (150, 70)
angle = -30
startAngle = 0
endAngle = 360
color = (0, 255, 0)
image = cv2.ellipse(img_copy, center_coordinates, axesLength,
           angle, startAngle, endAngle, color, thickness)
plt.imshow(image)
plt.show()
```

# Drawing polygons
To add a poligon we use the function `cv2.polylines`.

Syntax: `cv2.polylines(image, [pts], isClosed, color, thickness)`

Parameters:

`image`: It is the image on which circle is to be drawn.

`pts`: Array of polygonal curves.

`npts`: Array of polygon vertex counters.

`ncontours`: Number of curves.

`isClosed`: Flag indicating whether the drawn polylines are closed or not. If they are closed, the function draws a line from the last vertex of each curve to its first vertex.

`color`: It is the color of polyline to be drawn. For BGR, we pass a tuple.

`thickness`: It is thickness of the polyline edges.

```
img_copy = img.copy()
# Using cv2.polylines() method
# Draw a red polygon with thickness of 4 px

# Polygon corner points coordinates
pts = np.array([[25, 70], [25, 160],
                [110, 200], [200, 160],
                [200, 70], [110, 20]],
               np.int32)

#pts = pts.reshape((-1, 1, 2))

isClosed = True

color = (255, 0, 0)

# Line thickness of 4 px
thickness = 4

image = cv2.polylines(img_copy, [pts], isClosed, color, thickness)
plt.imshow(image)
plt.show()
```

Now let's add text on the image. The `cv2.putText` function of OpenCV is responsible for drawing text on an image. As a font example, let'a use `cv2.FONT_HERSHEY_SIMPLEX`. The available fonts are listed [here](https://docs.opencv.org/3.4.1/d0/de1/group__core.html#ga0f9314ea6e35f99bb23f29567fc16e11).

```
# Copy the image
img_copy = img.copy()
# Write text on the image
cv2.putText(img_copy, "Wall of love!!!", (200, 300), cv2.FONT_HERSHEY_SIMPLEX, 3.5, (255, 255, 0), 15)
plt.imshow(img_copy)
plt.show()
```

# Drawing an atom

```
# Create black empty image using np.zeros
W = 400
H = 400
size = H, W, 3
atom_image = np.zeros(size, dtype=np.uint8)
thickness = 2
line_type = 8

#draw ellipse with angle 90
angle=90
cv2.ellipse(atom_image,(W // 2, W // 2), (W // 4, W // 16),
                angle, 0, 360, (255, 0, 0), thickness, line_type)
angle=0
cv2.ellipse(atom_image,(W // 2, W // 2), (W // 4, W // 16),
                angle, 0, 360, (255, 0, 0), thickness, line_type)
angle=45
cv2.ellipse(atom_image,(W // 2, W // 2), (W // 4, W // 16),
                angle, 0, 360, (255, 0, 0), thickness, line_type)
angle=-45
cv2.ellipse(atom_image,(W // 2, W // 2), (W // 4, W // 16),
                angle, 0, 360, (255, 0, 0), thickness, line_type)

# Draw the nucleus
cv2.circle(atom_image, center=(200, 200), radius=W // 32, color=(0, 0, 255),
           thickness=-1)
plt.imshow(atom_image)
plt.show()
```

# Array slicing and cropping
Extracting “regions of interest” (ROIs) is an important skill for image processing. For now, let’s just manually extract an ROI. This can be accomplished with array slicing.

We can extract the ROI including "Ti amo" using Array slicing that follows the format: image[startY:endY, startX:endX]

```
# input image starting at x=750,y=450 at ending at x=850,y=490
roi = img[450:490, 750:850]
plt.imshow(roi)
plt.show()
```
