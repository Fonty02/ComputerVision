# Fontana Emanuele, Logaritmic Contrast and Sigmoid transformations #

```python
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt

image = "Dog.png"
img = cv2.imread(image, cv2.IMREAD_COLOR)
grayImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
lookUpTable = np.empty((1,256), dtype=np.uint8)
c = 3
for i in range(256):
    lookUpTable[0,i] =np.clip(c * np.log(1 + i), 0, 255)
print(lookUpTable)

#plo the function
x = np.arange(0, 256, 1)
y = c * np.log(1 + x)
plt.plot(x, y, color='blue')
plt.title('Logarithmic Function', fontsize=18)
plt.xlabel('Input Intensity', fontsize=14)
plt.ylabel('Output Intensity', fontsize=14)
plt.xlim([0, 255])
plt.ylim([0, 255])
plt.grid()
plt.show()

resultImage = cv2.LUT(grayImage, lookUpTable)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(grayImage, cmap='gray')
ax[0].set_title('Original', fontsize=18)
ax[0].set_xticks([])
ax[0].set_yticks([])
ax[1].imshow(resultImage, cmap='gray')
ax[1].set_title('Logarithmic Contrast', fontsize=18)
ax[1].set_xticks([])
ax[1].set_yticks([])
plt.tight_layout()
plt.show()
```

```python
lookUpTable = np.empty((1,256), np.uint8)
c = 100
y=2.6
for i in range(256):
    lookUpTable[0,i] =np.clip(round(1 / np.exp(-y * (i - c))),0,255)

#plot the function
x = np.arange(0, 256, 1)
y = 1 / np.exp(-y * (x - c))
plt.plot(x, y, color='blue')
plt.title('Sigmoid Function', fontsize=18)
plt.xlabel('Input Pixel Value', fontsize=14)
plt.ylabel('Output Pixel Value', fontsize=14)
plt.xlim(0, 255)
plt.ylim(0, 255)
plt.grid()
plt.show()

print(lookUpTable)
resultImage = cv2.LUT(grayImage, lookUpTable)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(grayImage, cmap='gray')
ax[0].set_title('Original', fontsize=18)
ax[0].set_xticks([])
ax[0].set_yticks([])
ax[1].imshow(resultImage, cmap='gray')
ax[1].set_title('Sigmoid', fontsize=18)
ax[1].set_xticks([])
ax[1].set_yticks([])
plt.tight_layout()
plt.show()
```
