# Fontana Emanuele Ex4 : Image denoising #

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Read the image and add noise
# 2. Aplly different denoising techniques to both of them and compare the results #

```python
img= cv2.cvtColor(cv2.imread('Dog.png'), cv2.COLOR_BGR2RGB)
rows,cols,_= img.shape
noise_percent = 0.3
num_pixel = int(rows * cols * noise_percent)
#create random coordinates to add noise
coords = (np.random.randint(0, rows, num_pixel), np.random.randint(0, cols, num_pixel))

img_noisy = img.copy()
for i in range(num_pixel // 2):
    img_noisy[coords[0][i], coords[1][i]] = [255, 255, 255]  # white dots
for i in range(num_pixel // 2, num_pixel):
    img_noisy[coords[0][i], coords[1][i]] = [0, 0, 0]  # black dots
plt.subplot(121), plt.imshow(img)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(img_noisy)
plt.title('Image with Noise'), plt.xticks([]), plt.yticks([])
plt.show()

#Mean Filter
kernel = np.ones((5, 5), np.float32) / 25
img_mean = cv2.filter2D(img, -1, kernel)
img_mean_noisy = cv2.filter2D(img_noisy, -1, kernel)
plt.subplot(121), plt.imshow(img_mean)
plt.title('Mean Filter'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(img_mean_noisy)
plt.title('Mean Filter on Noisy Image'), plt.xticks([]), plt.yticks([])
plt.show()

#Median Filter
img_median = cv2.medianBlur(img, 5)
img_median_noisy = cv2.medianBlur(img_noisy, 5)
plt.subplot(121), plt.imshow(img_median)
plt.title('Median Filter'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(img_median_noisy)
plt.title('Median Filter on Noisy Image'), plt.xticks([]), plt.yticks([])
plt.show()

#Gaussian Filter
img_gaussian = cv2.GaussianBlur(img, (5, 5), 0)
img_gaussian_noisy = cv2.GaussianBlur(img_noisy, (5, 5), 0)
plt.subplot(121), plt.imshow(img_gaussian)
plt.title('Gaussian Filter'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(img_gaussian_noisy)
plt.title('Gaussian Filter on Noisy Image'), plt.xticks([]), plt.yticks([])
plt.show()
```
