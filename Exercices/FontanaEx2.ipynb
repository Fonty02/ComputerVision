## Reading and visualizing an image ##

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img= cv2.imread('Dog.png')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

## Let's draw a 3x3 grid in green ##

```python
#let's save a copy of our image
copy_img= img.copy()
#let's split the image into 9 equal parts with a 3x3 grid
#we will use the height and width of the image to calculate the size of each part
height, width, _ = img.shape
part_height= height//3
part_width= width//3
#let's draw the grid with cv2 lines
for i in range(1,3):
    cv2.line(copy_img, (0, i*part_height), (width, i*part_height), (0, 255, 0), 10)
    cv2.line(copy_img, (i*part_width, 0), (i*part_width, height), (0, 255, 0), 10)
#let's show the image with the grid
plt.imshow(cv2.cvtColor(copy_img, cv2.COLOR_BGR2RGB))
plt.show()
```

## Arrows ##

```python
copy_img= img.copy()
cv2.arrowedLine(copy_img, (0, 0), (width//2, height//2), (0, 255, 0), 8, tipLength=0.1)
cv2.arrowedLine(copy_img, (0, height), (width//2, height//2), (255, 0, 0), 8, tipLength=0.1)
cv2.arrowedLine(copy_img, (width, 0), (width//2, height//2), (0, 0, 255), 8, tipLength=0.1)
cv2.arrowedLine(copy_img, (width, height), (width//2, height//2), (0, 0, 0), 8, tipLength=0.1)
plt.imshow(cv2.cvtColor(copy_img, cv2.COLOR_BGR2RGB))
plt.show()
```
